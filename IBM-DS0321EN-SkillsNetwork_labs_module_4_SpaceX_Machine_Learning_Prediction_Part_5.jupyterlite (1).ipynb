This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  12


Find the method performs best:


```python
from sklearn.model_selection import cross_val_score
import pandas as pd

# Set the number of folds for cross-validation
num_folds = 5

# Perform cross-validation for each model and store the scores in a dictionary
model_scores = {'knn_cv': cross_val_score(knn_cv, X_train, Y_train, cv=num_folds).mean(),
                'tree_cv': cross_val_score(tree_cv, X_train, Y_train, cv=num_folds).mean(),
                'svm_cv': cross_val_score(svm_cv, X_train, Y_train, cv=num_folds).mean(),
                'logreg_cv': cross_val_score(logreg_cv, X_train, Y_train, cv=num_folds).mean()}

# Convert the dictionary to a Pandas DataFrame
scores_df = pd.DataFrame.from_dict(model_scores, orient='index', columns=['cv_score'])

# Sort the DataFrame by score in descending order
scores_df = scores_df.sort_values(by='cv_score', ascending=False)

# Print the table of scores
print(scores_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score

# Set the number of folds for cross-validation
num_folds = 5

# Perform cross-validation for each model and store the scores in a dictionary
models = {'knn_cv': knn_cv, 
          'tree_cv': tree_cv, 
          'svm_cv': svm_cv, 
          'logreg_cv': logreg_cv}
model_scores = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, Y_train, cv=num_folds)
    model_scores[name] = {'cv_mean': np.mean(scores), 'cv_std': np.std(scores)}

# Convert the dictionary to a Pandas DataFrame
scores_df = pd.DataFrame.from_dict(model_scores, orient='index')

# Add a column for test scores using the models trained on X_train and Y_train
scores_df['test_score'] = [model.score(X_test, Y_test) for _, model in models.items()]

# Sort the DataFrame by cross-validation score in descending order
scores_df = scores_df.sort_values(by='cv_mean', ascending=False)

# Set up the plot
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(8, 6))

# Add a bar for the cross-validation scores with error bars
sns.barplot(x=scores_df.index, y='cv_mean', data=scores_df, ci='sd', capsize=0.1, color='blue', ax=ax)

# Add a bar for the test scores
sns.barplot(x=scores_df.index, y='test_score', data=scores_df, color='orange', ax=ax)

# Set the title and axis labels
ax.set_title('Cross-Validation and Test Scores for Classification Models')
ax.set_xlabel('Model')
ax.set_ylabel('Score')
ax.set_ylim(0, 1)

# Add a legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Cross-Validation', 'Test'], loc='upper right')

plt.show()
```
